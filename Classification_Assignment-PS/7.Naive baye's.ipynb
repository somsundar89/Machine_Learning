```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset["classification_yes"].value_counts()
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep=dataset['classification_yes']
```

```python
dep.shape
```

```python
dep
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_= sc.fit_transform(X_train)
X_test_= sc.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, CategoricalNB
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# Create a pipeline with a placeholder for Naive Bayes
pipe = Pipeline([
    ('nb', MultinomialNB())  # placeholder, will be replaced in grid
])

# Define parameter grid
param_grid = [
    {
        'nb': [MultinomialNB()],
        'nb__alpha': [0.1, 0.5, 1.0]
    },
    {
        'nb': [BernoulliNB()],
        'nb__alpha': [0.1, 0.5, 1.0],
        'nb__binarize': [0.0, 0.5]
    },
    {
        'nb': [CategoricalNB()],
        'nb__alpha': [0.1, 0.5, 1.0]
    }
]

# GridSearchCV
grid = GridSearchCV(pipe, param_grid, refit=True, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)

print("Best Estimator:", grid.best_estimator_)
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)

# Predict using the best model
y_pred = grid.predict(X_test)
```

```python
re=grid.cv_results_
#print(re)
grid_predictions = grid.predict(X_test_) 
   

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)



# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```
