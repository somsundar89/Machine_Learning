```python
#Importing pandas library to access and work with tabular data like excel, csv.

import pandas as pd 
```

```python
# In this part related to load the csv data from our directory to understand the input and output data
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# For my understanding, Im validating the data is correct and check whether the csv's were assigned to the dataset Variable or not.
dataset
```

```python
# Now, Input / Output split has been taken place

independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]]
```

```python
# check whether the splitation has been performed or not (Input variable)
independent
```

```python
# check whether the splitation has been performed or not (Output Variable)
dependent
```

```python
#Splitting Train set and test set using sklearn library -- Inside we have train_test_split method to perform this task
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation under the LinearRegression Class. In the name of fit refers to method of creating model with our training set (in this formation y=mX+c)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
LinearRegression()
```

```python
# Once training model has been created, we need to validate the weight has been calculated (Here weight is mass, slope, constant distance b/w to two datapoints)
weight=regressor.coef_
```

```python
# validate the weight output

weight
```

```python
# Here Bais is calculated using intercept_ term. (This is intercept value, starting point of straight line or minimum value)
bais=regressor.intercept_
```

```python
# validate the bais output

bais
```

```python
#Once the Model is created we need to test this using predicted output, using test input will predict and save in y_pred variable
y_pred=regressor.predict(X_test)
```

```python
# Using r2_score method, Evaluate the predicted output and test output. Based on the score we will conclude whether the created model is good or not

from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
# If the value Nearing to 1, the model is good., Nearing to 0, then the model is bad 
r_score
```

```python
# Using Pickle library we can save the model for our deployment phase
import pickle
```

```python
# create the file name for our model with the extension of .sav., 
filename="Finalized_Simple_LinearModel.sav"
```

```python
# here dump refers to dumping our created model into this filename.sav. with write binary attribute
pickle.dump(regressor,open("Finalized_Simple_LinearModel.sav",'wb'))
```

```python
#Successfully dumped.., Now we need to load and check the model for our deployment phase
loaded_model=pickle.load(open("Finalized_Simple_LinearModel.sav",'rb'))
```

```python
#Get inputs for prediction using the loaded model
result=loaded_model.predict([[15]])
```
